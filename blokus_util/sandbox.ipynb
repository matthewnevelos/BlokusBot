```python
class Shape:
    def __init__(self, name, color):
        self.name = name
        self.color = color

    def __str__(self):
        return self.name
    
    def __repr__(self):
        return self.name

# Example usage
shapes = [Shape("Circle", "Red"), Shape("Square", "Blue"), Shape("Triangle", "Green")]
print(shapes)
```
